```python
import re
def remove_url(txt):
    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())
```

```python
#creamos una lista de listas que contengan las palabas en minusculas para cada tweet
words_in_tweet = [tweet.lower().split() for tweet in all_tweets_no_url]
words_in_tweet[:2]
```

```python
#vemos la frecuencia de cada palabra
#conda install -c conda-forge itertools
#conda install -c conda-forge textblob
#conda install -c conda-forge twython
import itertools
import collections #crea colecciones como un dict especial
#si no existe la palabra la agrega al dict
# Lista de todas las palabras en todos los tweets
all_words_no_urls = list(itertools.chain(*words_in_tweet))
#Cuenta las palabras
counts_no_urls = collections.Counter(all_words_no_urls)
#Selecciona las 15 mas utilizadas
counts_no_urls.most_common(15)#metodo de collection.Counter
```

```python
all_words_no_urls
```

```python
#Utilizamos pandas
clean_tweets_no_urls = pd.DataFrame(counts_no_urls.most_common(15),columns=['words','count'])
clean_tweets_no_urls.head()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8,8))
#Grafica de barra horizontal
clean_tweets_no_urls.sort_values(by='count').plot.barh(x='words', y='count',ax = ax,color="purple")
ax.set_title("Palabras mas utilizadas en tweets (search_words=#COVID)")
plt.show
```

# Sentiment Analysis
Se utiliza para identificar temas de interes en un rango de +1 a -1. Donde +1 es de mayor interes o de mayor positividad -1 es de menos interes, mayor negatividad

TextBlob se utiliza para calcular los valores de polaridad en tweets. Al crear objetos TextBlob se puede identificar el valor de polaridad por medio del metodo polarity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
import collections
#conda install -c conda-forge textblob
from textblob import TextBlob
```

```python
#Cambio climatico
search_item = "#climatechange -filter:retweets"
tweets = tw.Cursor(api.search_tweets, q=search_term, lang="en", since='2020-01-01').items(1000)
#Remove URLs
tweets_no_urls = [remove_url(tweet.text) for tweet in tweets]
```

```python
#Analisis de sentimientos en los tweets
#Creamos objetos textlob
sentiment_objects = [TextBlob(tweet) for tweet in tweets_no_urls]
sentiment_objects[0].polarity, sentiment_objects[0]
```

```python
#Creamos una lista de polaridad por cada tweet
sentiment_values = [[tweet.sentiment.polarity, str(tweet)] for tweet in sentiment_objects]
sentiment_values
```

```python
sentiment_df = pd.DataFrame(sentiment_values, columns=["polarity","tweet"])
sentiment_df.head()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sentiment_df.hist(bins=[-1, -0.75, -0.5, -0.25, 0.25, 0.5, 0.75, 1], ax = ax, color = "cyan")
plt.title("Sentimientos de los tweets acerca del cambio climatico")
plt.show()
```

```python
#Podemos eliminar aquellos con polaridad igual a cero
sentiment_df = sentiment_df[sentiment_df.polarity != 0]
```

```python
#Grafica sin tomar en cuenta la polaridad igual a cero = 0.0
fig, ax = plt.subplots(figsize = (8,6))
sentiment_df.hist(bins=[-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1], ax = ax, color = "cyan")
plt.title("Sentimientos de los tweets acerca del cambio climatico")
plt.show()
```
